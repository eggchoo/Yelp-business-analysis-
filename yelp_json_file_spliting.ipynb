## splitting json files into multiple files

```python
import json 

input_file = r'C:\Users\08042\Desktop\projects\yelp project\Yelp dataset\yelp_academic_dataset_review.json'
output_prefix ='split_file_'
num_files = 10

# count total number of lines
with open(input_file, 'r', encoding='utf8') as f:
    total_lines = sum(1 for _ in f)

lines_per_file = total_lines // num_files 
print(f'Total lines: {total_lines}, Lines per file: {lines_per_file}')

# Split the file
with open(input_file, 'r', encoding='utf8') as f:
    for i in range(num_files):
        output_file = f'{output_prefix}{i+1}.json'
        with open(output_file, 'w', encoding='utf8') as outfile:
            for j in range(lines_per_file):
                line = f.readline()
                if not line:
                    break 
                outfile.write(line)
print('Splitting completed.')
```
